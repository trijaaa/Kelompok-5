numpy

```python
import numpy as np
```

```python
#membuat matrix 2x3 menggunakan numpy
A = np.array([[5, 2, 1], [6, 4, 2]])
A
```

```python
#membuat vektor 3x1 menggunakan numpy
b = np.array([4, 1, 2])
b
```

pytorch

```python
import torch
```

```python
#membuat matrix 2x3 menggunakan torch
A_pt = torch.tensor([[5, 2, 1], [6, 4, 2]])
A_pt
```

```python
#membuat vektor 3x1 menggunakan torch
b_pt = torch.tensor([4, 1, 2])
b_pt
```

tensorflow

```python
import tensorflow as tf
```

```python
#membuat matrix 2x3 menggunakan tensorflow
A_tf = tf.Variable([[5, 2, 1], [6, 4, 2]])
A_tf
```

```python
#membuat vektor 3x1 menggunakan tensorflow
b_tf = tf.Variable([4, 1, 2])
b_tf
```

```python
#membuat matrix B 3x3 menggunakan numpy
B = np.array([[3, 4, 2], [2, 1, 7], [5, 2, 1]])
B
```

```python
#membuat matrix B 3x3 menggunakan pytorch
B_pt = torch.tensor([[3, 4, 2], [2, 1, 7], [5, 2, 1]])
B_pt
```

```python
#membuat matrix B 3x3 menggunakan tensorflow
B_tf = tf.Variable([[3, 4, 2], [2, 1, 7], [5, 2, 1]])
B_tf
```

```python
#mengalikan matrix A dengan matrix B
np.dot(A, B)
```

```python
#mengalikan matrix A_pt dengan matrix B_pt
torch.matmul(A_pt, B_pt)
```

```python
tf.matmul(A_tf, B_tf)
```
