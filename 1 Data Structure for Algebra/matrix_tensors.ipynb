**SOAL NO 1 (Matrix Tensors)**

```python
import numpy as np
```

buat matrix X berukuran 2x3 menggunakan numpy

```python
X = np.array([[15, 7, 20], [10, 6, 4]])
print(X)
```

tampilkan ukuran dari matrix X

```python
print(X.shape)
```

tampilkan seluruh elemen dari kolom pertama dari matrix X

```python
print(X[:,0])
```

tampilkan seluruh elemen dari baris kedua dari matrix X

```python
print(X[1,:])
```

**SOAL NO 2 (Matrix Tensors) (PyTorch)**

ulangi soal nomor 1 menggunakan pytorch

```python
import torch

X_pt = torch.tensor([[15, 7, 20], [10, 6, 4]])
print(X_pt)
print()

print(X_pt.shape)
print()

print(X_pt[:,0])
print()

print(X_pt[1,:])
```

ulangi soal nomor 1 menggunakan tensorflow

```python
import tensorflow as tf

X_tf = tf.Variable([[15, 7, 20], [10, 6, 4]])
print(X_tf)
print()

print(tf.shape(X_tf))
print()

print(X_tf[:,0])
print()

print(X_tf[1,:])
```
