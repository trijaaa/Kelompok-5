**SOAL NO 2 (Norms and Unit Vectors)**

buat vector v dengan elemen (3,4)

```python
v = np.array([3, 4])
print(f'Vektor asli: {v}')
```

hitung nilai L2 norm dari vector v

```python
norm = np.linalg.norm(v)
print(f'Panjang vektor: {norm}')
```

tampilkan unit vector dari vector v

```python
unit_v = v / norm
print(f'Unit vektor: {unit_v}')
```

hitung panjang unit vector dari vector v

```python
print(f'Panjang unit vektor: {np.linalg.norm(unit_v)}')
```
