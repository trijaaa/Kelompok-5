**SOAL NO 1 (VECTORS AND VECTOR TRANSPOSITION)**

buat vector x dengan ukuran 3x1

```python
import numpy as np

x = np.array([[10],
              [15],
              [20]])
print(x)
```

hitung panjang vector x

```python
print(f'Panjang vector x: {len(x)}')
```

tampilkan ukuran vector x

```python
print(f'Ukuran dari vector x: {x.shape}')
```

tampilkan tipe data dari vector x

```python
print(f'Tipe dari vector x: {type(x)}')
```

akses elemen pertama dari vector x

```python
print(f'Elemen pertama dari vector x: {x[0]}')
```

tampilkan tipe data elemen pertama dari vector x

```python
print(f'Tipe data elemen pertama vector x: {type(x[0])}')
```

**SOAL NO 2 (VECTORS AND VECTOR TRANSPOSITION)**

lakukan transpose pada vector x dan simpan pada variabel x_t

```python
x_trans = x.transpose()
print('Transpose dari vector x adalah:')
print(x_trans)
```

tampilkan ukuran dari vector x_t

```python
print('Ukuran dari vector x_t:')
print(x_trans.shape)
```

lakukan transpose pada vector y dan simpan pada variabel y_t

```python
y = np.array([[10],
              [15],
              [20]])
print(y)
```

tampilkan ukuran dari vector x_t

```python
print(f'Ukuran dari vector y: {y.shape}')
```

**SOAL NO 3 (VECTORS AND VECTOR TRANSPOSITION) (TENSORFLOW)**

lakukan soal nomor 2 menggunakan pytorch dan tensorflow

```python
import tensorflow as tf

x_tf = tf.Variable([[10],
                    [15],
                    [20]])
print(x_tf)
```

```python
print(x_tf.shape)
```

```python
y_tf = tf.Variable([[10],
                    [15],
                    [20]])
print(y_tf)
```

```python
print(y_tf.shape)
```

**SOAL NO 3 (VECTORS AND VECTOR TRANSPOSITION) (PYTORCH)**

```python
print(x_pt.shape)
```

```python
y_pt = torch.tensor([[10],
                    [15],
                    [20]])
print(y_pt)
```

```python
print(y_pt.shape)
```
