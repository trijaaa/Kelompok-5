**SOAL NO 2 (SCALAR) (TENSORFLOW)**

lakukan tahapan pada soal 1 dengan menggunakan tensorflow

```python
import tensorflow as tf

x_tf = tf.Variable(25)
```

tampilkan tipe data dari x_tf (Tensorflow)

```python
print(f'Type dari x_tf adalah: {type(x_tf)}')
```

buat variabel skalar y_tf dan simpan suatu nilai skalar (Tensorflow)

```python
y_tf = tf.Variable(50)
```

tampilkan tipe data dari y_tf (Tensorflow)

```python
print(f'Type dari y_tf adalah: {type(y_tf)}')
```

jumlahkan x_tf + y_tf (Tensorflow)

```python
penjumlahan3 = x_tf + y_tf
print(penjumlahan3)
```

buat variabel bernama x_float_tf dan simpan suatu nilai desimal (Tensorflow)

```python
x_float_tf = tf.Variable(3.90)
```

jumlahkan x_float_tf dengan y_tf (Tensorflow)

```python
y_float_tf = float(y_tf)
penjumlahan4 = x_float_tf + y_float_tf
print(penjumlahan4)
```

tampilkan tipe data dari x_float_tf (Tensorflow)

```python
print(f'Type dari x_float_tf adalah: {type(x_float_tf)}')
```

**SOAL NO 3 (SCALAR) (PYTORCH)**

lakukan tahapan pada soal 1 dengan menggunakan pytorch

```python
import torch

x_pt = torch.tensor(25)
```

tampilkan tipe data dari x_pt (PyTorch)

```python
print(f'Type dari x_pt adalah: {type(x_pt)} dengan nilai {x_pt}')
```

```python
y_pt = torch.tensor(50)

print(f'Type dari y_pt adalah: {type(y_pt)} dengan nilai {y_pt}')
```

```python
penjumlahan5 = x_pt + y_pt
print(penjumlahan5)
```

```python
x_float_pt = torch.tensor(3.90)

penjumlahan6 = x_float_pt + y_pt
print(penjumlahan6)
```

```python
print(f'Type dari x_float_pt adalah: {type(x_float_pt)} dengan nilai {x_float_pt}')
```
