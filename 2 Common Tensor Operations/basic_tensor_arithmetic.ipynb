```python
import numpy as np
import torch
import tensorflow as tf
```

SOAL NO 1 (BASIC TENSOR ARITHMETIC (THE HADAMARD PRODUCT))

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
print("Matrix X dengan ukuran 2x3 menggunakan numpy:")
X = np.array(([5, 15, 25], [35, 45, 55]))
print(X)
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor(([5, 15, 25], [35, 45, 55]))
print("Matrix X_pt dengan ukuran 2x3 menggunakan pytorch:")
print(X_pt)
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable(([5, 15, 25], [35, 45, 55]))
print("Matrix dengan ukuran 2x3 menggunakan tensorflow:")
print(X_tf)
```

lakukan perkalian matrix X, X_pt, X_tf dengan suatu skalar

```python
# Perkalian X dengan suatu skalar
print("Hasil perkalian X dengan suatu skalar:")
print(X * 2)
```

```python
# Perkalian X_pt dengan suatu skalar
print("Hasil perkalian X_pt dengan suatu skalar:")
print(torch.mul(X_pt, 2))
```

```python
# Perkalian X_tf dengan suatu skalar
print("Hasil perkalian X_tf dengan suatu skalar:")
print(tf.multiply(X_tf, 2))
```

lakukan penjumlahan matrix X, X_pt, X_tf dengan suatu skalar

```python
# Penjumlahan X dengan suatu skalar
print("Hasil penjumlahan X dengan suatu skalar:")
print(X + 2)
```

```python
# Penjumlahan X_pt dengan suatu skalar
print("Hasil penjumlahan X_pt dengan suatu skalar:")
print(torch.add(X_pt, 2))
```

```python
# Penjumlahan X_tf dengan suatu skalar
print("Hasil penjumlahan X_tf dengan suatu skalar:")
print(tf.add(X_tf, 2))
```

lakukan pengurangan matrix X, X_pt, X_tf dengan suatu skalar

```python
# Pengurangan X dengan suatu skalar
print("Hasil pengurangan X dengan suatu skalar:")
print(X - 2)
```

```python
# Pengurangan X_pt dengan suatu skalar
print("Hasil pengurangan X_pt dengan suatu skalar:")
print(torch.sub(X_pt, 2))
```

```python
# Pengurangan X_tf dengan suatu skalar
print("Hasil pengurangan X_tf dengan suatu skalar:")
print(tf.subtract(X_tf, 2))
```

SOAL NO 2 (BASIC TENSOR ARITHMETIC (THE HADAMARD PRODUCT))

```python
# buat matrix Y dengan ukuran 2x3 menggunakan numpy
Y = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix Y dengan ukuran 2x3 menggunakan numpy:")
print(Y)
```

```python
# buat matrix Y_pt dengan ukuran 2x3 menggunakan pytorch
Y_pt = torch.tensor([[1, 2, 3], [4, 5, 6]])
print("Matrix Y_pt dengan ukuran 2x3 menggunakan numpy:")
print(Y_pt)
```

```python
# buat matrix Y_tf dengan ukuran 2x3 menggunakan tensorflow
Y_tf = tf.Variable([[1, 2, 3], [4, 5, 6]])
print("Matrix Y_tf dengan ukuran 2x3 menggunakan numpy:")
print(Y_tf)
```

```python
# kalikan X dengan Y
print("Hasil perkalian X dengan Y:")
print(X * Y)
```

```python
# kalikan X dengan Y
print("Hasil perkalian X dengan Y:")
print(X * Y)
```

```python
# kalikan X_pt dengan Y_pt
print("Hasil perkalian X_pt dengan Y_pt:")
print(torch.mul(X_pt, Y_pt))
```

```python
# kalikan X_tf dengan Y_tf
print("Hasil perkalian X_tf dengan Y_tf:")
print(tf.multiply(X_tf, Y_tf))
```
