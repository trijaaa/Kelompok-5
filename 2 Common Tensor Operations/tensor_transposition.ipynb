```python

import numpy as np
import torch
import tensorflow as tf

print("SOAL NO 1 (TENSOR TRANSPOSITION)")
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[10, 20, 30], [40, 50, 60]])
print("Matrix X dengan ukuran 2x3:")
print(X)
```

```python
# buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[10, 20, 30], [40, 50, 60]])
print("Matrix X_pt dengan ukuran 2x3:")
print(X_pt)
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[10, 20, 30], [40, 50, 60]])
print("Matrix X_tf dengan ukuran 2x3:")
print(X_tf)
```

```python
print("SOAL NO 2 (TENSOR TRANSPOSITION)")
```

```python
# lakukan transpose pada matrix X, X_pt, X_tf
print("Transpose dari X:")
print(X.T)
print()

print("Transpose dari X_pt:")
print(X_pt.T)
print()

print("Transpose dari X_tf:")
print(tf.transpose(X_tf))
```
