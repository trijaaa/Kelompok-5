**SOAL NO 1 (TENSOR REDUCTION)**

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array(([2, 12, 22], [32, 42, 52]))
print("Matrix X dengan ukuran 2x3 menggunakan numpy:")
print(X)
```

```python
X.sum()
```

```python
# matrix X_pt ukuran 2x3 pytorch
X_pt = torch.tensor(([2, 12, 22], [32, 42, 52]))
print("Matrix X_pt dengan ukuran 2x3 menggunakan pytorch:")
print(X_pt)
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable(([2, 12, 22], [32, 42, 52]))
print("Matrix X_tf dengan ukuran 2x3 menggunakan tensorflow:")
print(X_tf)
```

**SOAL NO 2 (TENSOR REDUCTION)**

```python
# hitung hasil penjumlahan seluruh elemen matrix X
print("Hasil penjumlahan seluruh elemen matrix X:")
print(X.sum())
```

```python
# hitung hasil penjumlahan seluruh elemen matrix X_pt
print("Hasil penjumlahan seluruh elemen matrix X_pt:")
print(torch.sum(X_pt))
```

```python
# hitung hasil penjumalahn seluruh elemen matrix X_tf
print("Hasil penjumlahan seluruh elemen matrix X_tf:")
print(tf.reduce_sum(X_tf))
```

**SOAL NO 3 (TENSOR REDUCTION)**

```python
# hitung hasil penjumlahan matrix X berdasarkan row nya
print("Hasil penjumlahan matrix X berdasarkan row nya:")
print(X.sum(axis=1))
```

```python
# hitung hasil penjumlahan matrix X_pt berdasarkan row nya
print("Hasil penjumlahan matrix X_pt berdasarkan row nya:")
print(torch.sum(X_pt, 1))
```

```python
# hitung hasil penjumlahan matrix X_tf berdasarkan row nya
print("Hasil penjumlahan matrix X_tf berdasarkan row nya:")
print(tf.reduce_sum(X_tf, 1))
```

**SOAL NO 4 (TENSOR REDUCTION)**

```python
# hitung hasil penjumlahan matrix X berdasarkan kolom nya
print("Hasil penjumlahan matrix X berdasarkan kolom nya:")
print(X.sum(axis=0))
```

```python
# hitung hasil penjumlahan matrix X_pt berdasarkan kolom nya
print("Hasil penjumlahan matrix X_pt berdasarkan kolom nya:")
print(torch.sum(X_pt, 0))
```

```python
# hitung hasil penjumlahan matrix X_tf berdasarkan row nya
print("Hasil penjumlahan matrix X_tf berdasarkan row nya:")
print(tf.reduce_sum(X_tf, 0))
```

**SOAL NO 5 (TENSOR REDUCTION)**

hitung nilai minimal dari matrix X, X_pt dan X_tf

```python
# nilai rata-rata dari matrix X
print(X.mean())
```

```python
# nilai rata-rata dari matrix X_tf
print(tf.reduce_mean(X_tf))
```
