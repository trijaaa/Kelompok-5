**SOAL NO 1 (THE DOT PRODUCT)**

```python
import numpy as np
import torch
import tensorflow as tf
```

buat vector x dan y dengan ukuran 1x3 menggunakan numpy

```python
x = np.array([5, 15, 25])
y = np.array([10, 20, 30])
print("Vector x dengan ukuran 1x3:")
print(x)
print()

print("Vector y dengan ukuran 1x3:")
print(y)
```

buat vector x_pt dan y_pt dengan ukuran 1x3 menggunakan pytorch

```python
x_pt = torch.tensor([5, 15, 25])
y_pt = torch.tensor([10, 20, 30])
print("Vector x_pt dengan ukuran 1x3:")
print(x_pt)
print()

print("Vector y_pt dengan ukuran 1x3:")
print(y_pt)
```

buat vector x_tf dan y_tf dengan ukuran 1x3 menggunakan tensorflow

```python
x_tf = tf.Variable([5, 15, 25])
y_tf = tf.Variable([10, 20, 30])
print("Vector x_tf dengan ukuran 1x3:")
print(x_tf)
print()

print("Vector y_tf dengan ukuran 1x3:")
print(y_tf)
```

**SOAL NO 2 (THE DOT PRODUCT)**

hitung perkalian dot product antara vector x dan vector y

```python
print("Hasil perkalian dot product antara vector x dan y:")
print(np.dot(x, y))
```

hitung perkalian dot product antara vector x_pt dan vector y_pt

```python
print("Hasil perkalian dot product antara vector x_pt dan y_pt:")
print(torch.dot(x_pt, y_pt))
```

hitung perkalian dot product antara vector x_tf dan vector y_tf

```python
print("Hasil perkalian dot product antara x_tf dan y_tf:")
print(tf.reduce_sum(tf.multiply(x_tf, y_tf)))
```

1. What is ***Y*** Transpose?

```python
Y = np.array(([42, 4, 7, 99], [-99, -3, 17, 22]))
Y_tp = Y.transpose()
print("Transpose dari Y adalah:")
print(Y_tp)
```

2. What is the Hadamard product of these matrices?

```python
X = np.array([[25, 10], [-2, 1]])
Y = np.array([[-1, 7], [10, 8]])
print("Hadamard product dari matriks tersebut adalah:")
print(X * Y)
```

3. What is the dot product of the tensors ***w*** and ***x***?

```python
w = np.array([-1, 2, -2])
x = np.array([5, 10, 0])
print("Dot product dari tensor w dan x adalah:")
print(np.dot(w, x))
```
