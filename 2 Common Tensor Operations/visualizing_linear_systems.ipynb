**SOAL NO 1 (VISUALIZING LINEAR SYSTEMS)**

buat visualisasi dari persamaan berikut 
y = 2x+1 dan 3x-y = 5

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = 2*x + 1
y2 = 3*x - 5

x_titik = 6
y_titik = 13

fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')

plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-2, 10])
ax.set_ylim([-6, 20])

ax.plot(x, y1, c='green', label='y = 2x + 1')
ax.plot(x, y2, c='brown', label='y = 3x - 5')

plt.axvline(x=6, color='purple', linestyle='--')
plt.axhline(y=13, color='purple', linestyle='--')

ax.scatter(x_titik, y_titik)

plt.legend()
plt.show()
```
