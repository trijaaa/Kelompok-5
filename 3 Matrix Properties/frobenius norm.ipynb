```python
import numpy as np
```

```python
#membuat matrix 3x3 dengan numpy
X = np.array([[6, 3, 1], [6, 4, 2], [8, 5, 7]])
X
```

```python
import torch
```

```python
#membuat matrix 3x3 dengan pytorch
X_pt = torch.tensor([[6.0, 3.0, 1.0], [6.0, 4.0, 2.0], [8.0, 5.0, 7.0]])
```

```python
#menghitung nilai frobenius norm
torch.norm(X_pt)
```

Tensorflow

```python
import tensorflow as tf
```

```python
#membuat matrix 3x3 dengan tensorflow
X_tf = tf.Variable([[6.0, 3.0, 1.0], [6.0, 4.0, 2.0], [8.0, 5.0, 7.0]])
```

```python
#menghitung nilai frobenius norm
tf.norm(X_tf)
```
