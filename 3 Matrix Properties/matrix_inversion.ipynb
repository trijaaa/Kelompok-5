**SOAL NO 1 (MATRIX INVERSION)**

```python
import numpy as np
import torch
import tensorflow as tf
```

buat matrix X ukuran 2x2 menggunakan numpy

```python
X = np.array(([12, 20], [9, 30]))
print(X)
```

buat invers dari matrix X dan beri nama Xinv menggunakan numpy

```python
Xinv = np.linalg.inv(X)
Xinv
```

kalikan X dengan Xinv menggunakan numpy

```python
np.dot(Xinv, X)
```

**SOAL NO 2 (MATRIX INVERSION)**

selesaikan persamaan berikut menggunakan numpy, tensorflow dan pytorch

![soal](image.png)

**NUMPY**

```python
X = np.array([[4, 2], [-5, -3]])
Y = np.array([[4], [-7]])
Z = np.linalg.solve(X, Y)
print(Z)
```

**TORCH**

```python
X_pt = torch.tensor([[4, 2], [-5, -3]], dtype=float)
Y_pt = torch.tensor([[4], [-7]], dtype=float)
Z_pt = torch.linalg.solve(X_pt, Y_pt)
print(Z_pt)
```

**TENSORFLOW**

```python
X_tf = tf.Variable([[4, 2], [-5, -3]], dtype=tf.float64)
Y_tf = tf.Variable([[4], [-7]], dtype=tf.float64)
Z_tf = tf.linalg.solve(X_tf, Y_tf)
print(Z_tf)
```
